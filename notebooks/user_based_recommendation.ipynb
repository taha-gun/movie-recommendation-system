```python
def user_based_recommender(target_user, user_movie_matrix, similarity_threshold=0.30, min_rating_count=2, top_n=10):
    # NaN'ları 0 yap
    user_movie_filled = user_movie_matrix.fillna(0)

    # cosine similarity
    user_similarity = cosine_similarity(user_movie_filled)

    user_similarity_df = pd.DataFrame(
        user_similarity,
        index=user_movie_filled.index,
        columns=user_movie_filled.index
    )

    # benzer kullanıcıları bul
    similar_users = user_similarity_df[target_user].sort_values(ascending=False)

    # kendisini çıkar
    similar_users = similar_users.drop(index=target_user)

    # threshold uygula
    similar_users = similar_users[similar_users > similarity_threshold]

    if similar_users.empty:
        return "Bu threshold ile benzer kullanıcı bulunamadı."

    # hedef kullanıcının izlediği filmler
    target_user_ratings = user_movie_matrix.loc[target_user]
    rated_movies = target_user_ratings[target_user_ratings.notna()].index

    # izlemediği filmler
    unrated_movies = user_movie_matrix.columns.difference(rated_movies)

    # benzer kullanıcıların ratingleri
    similar_users_ratings = user_movie_matrix.loc[similar_users.index]

    # similarity dataframe
    similar_users_df = similar_users.reset_index()
    similar_users_df.columns = ["userId", "similarity"]

    # long format
    similar_users_ratings_long = similar_users_ratings.stack().reset_index()
    similar_users_ratings_long.columns = ["userId", "title", "rating"]

    # similarity merge
    similar_users_ratings_long = similar_users_ratings_long.merge(similar_users_df, on="userId")

    # weighted rating
    similar_users_ratings_long["weighted_rating"] = (
        similar_users_ratings_long["rating"] * similar_users_ratings_long["similarity"]
    )

    # sadece hedef kullanıcının izlemediği filmler
    candidate_movies = similar_users_ratings_long[
        similar_users_ratings_long["title"].isin(unrated_movies)
    ]

    if candidate_movies.empty:
        return "Aday film bulunamadı."

    # skor hesapla
    recommendation_scores = candidate_movies.groupby("title").agg({
        "weighted_rating": "sum",
        "similarity": "sum",
        "rating": "count"
    })

    recommendation_scores["recommendation_score"] = (
        recommendation_scores["weighted_rating"] / recommendation_scores["similarity"]
    )

    recommendation_scores = recommendation_scores.rename(columns={"rating": "rating_count"})

    # minimum oy filtresi
    recommendation_scores = recommendation_scores[
        recommendation_scores["rating_count"] >= min_rating_count
    ]

    if recommendation_scores.empty:
        return "Filtrelerden sonra öneri kalmadı."

    # sırala
    recommendation_scores = recommendation_scores.sort_values(
        by="recommendation_score", ascending=False
    )

    return recommendation_scores.head(top_n)
```

```python
user_based_recommender(target_user=5, user_movie_matrix=user_movie_matrix)
```
